```python
import pickle
import json
import os

# Load event2word and word2event from dataset\pre\dictionary.pkl
with open(os.path.join('pre', 'dictionary.pkl'), 'rb') as f:
    event2word, word2event = pickle.load(f)

# Save event2word and word2event as JSON files for inspection
with open('event2word.json', 'w') as f:
    json.dump(event2word, f, indent=4)
with open('word2event.json', 'w') as f:
    json.dump(word2event, f, indent=4)
```
